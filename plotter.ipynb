```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

% matplotlib inline
```

```python
df = pd.read_csv('test_data/rna_and_peptide_counts.csv')
df
```

```python
sorted_index = df['rna_con_match'].sort_values().index
sorted_index
```

```python
sorted_index = df['rna_con_match'].sort_values().index

fig = plt.figure(figsize=(10, 5))
plt.title('RNA Statistics Ordered by RNA Matching Consensus')
plt.xlabel('Sample')
plt.ylabel('Count')
plt.scatter(range(len(sorted_index)), df.iloc[sorted_index]['length'].values, label='Consensus length')
plt.scatter(range(len(sorted_index)), df.iloc[sorted_index]['rna_con_match'].values, label='Consensus with RNA')
plt.scatter(range(len(sorted_index)), df.iloc[sorted_index]['rna_con_mismatch'].values, label='Non Consensus with RNA')
plt.scatter(range(len(sorted_index)), df.iloc[sorted_index]['con_no_rna'].values, label='Consensus without RNA')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
sorted_index = df['rna_con_mismatch'].sort_values().index

fig = plt.figure(figsize=(10, 5))
plt.title('RNA Statistics by Mismatch')
plt.xlabel('Sample')
plt.ylabel('Count')
plt.scatter(range(len(sorted_index)), df.iloc[sorted_index]['length'].values, label='Consensus length')
plt.scatter(range(len(sorted_index)), df.iloc[sorted_index]['rna_con_match'].values, label='Consensus with RNA')
plt.scatter(range(len(sorted_index)), df.iloc[sorted_index]['rna_con_mismatch'].values, label='Non Consensus with RNA')
plt.scatter(range(len(sorted_index)), df.iloc[sorted_index]['con_no_rna'].values, label='Consensus without RNA')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
sorted_index = df['aa_with_rna'].sort_values().index
sorted_index
```

```python
sorted_index = df['total_aa'].sort_values().index

fig = plt.figure(figsize=(10, 5))
plt.title('Amino Acids Statistics Ordered by Total Number of Amino Acids')
plt.xlabel('Sample')
plt.ylabel('Count')
plt.scatter(range(len(sorted_index)), df.iloc[sorted_index]['total_aa'].values, label='Total AA')
plt.scatter(range(len(sorted_index)), df.iloc[sorted_index]['aa_with_rna'].values, label='AA with RNA')
plt.scatter(range(len(sorted_index)), df.iloc[sorted_index]['aa_without_rna'].values, label='AA without RNA')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
